```python
import pandas as pd
import altair as alt
```

```python
alt.data_transformers.disable_max_rows()
```

```python
data = pd.read_excel("2014-2020.xlsx")
data.head()
```

```python
daily_data = data[data['Час/Дата'].str[:3]=="24-"]
daily_data["date"] = pd.to_datetime(daily_data['Час/Дата'].str[3:], format='%d.%m.%Y')


weekly_data = daily_data[daily_data['Час/Дата'].str[3:5].astype('int32') % 7 == 1]
weekly_data["date"] = pd.to_datetime(weekly_data['Час/Дата'].str[3:], format='%d.%m.%Y')
weekly_data = weekly_data.drop(columns=['Час/Дата'])


monthly_data = data[data['Час/Дата'].str[:5]=="24-01"]
monthly_data["date"] = pd.to_datetime(monthly_data['Час/Дата'].str[3:], format='%d.%m.%Y')
monthly_data = monthly_data.drop(columns=['Час/Дата'])

daily_data = daily_data.drop(columns=['Час/Дата'])

monthly_data.head()
```

```python
daily_data.head()
```

```python
weekly_data.head()
```

```python
sources = ["AES", "TEC", "VDE", "TES", "GES", "GAES_GEN"]
data_by_source = []

for source in sources:
    temp_data = weekly_data[['date', source]].rename(columns={source: "production"})
    temp_data["source"] = source
    data_by_source.append(temp_data)
    
working_data = pd.concat(data_by_source, ignore_index=True)
print(working_data["source"].unique())

working_data["order"] = 0 
for source in sources:
    current_subset = working_data['source'] == source
    working_data['order'] += (current_subset).astype('int32') * working_data[current_subset]['production'].mean()
    print(source, working_data[current_subset]['production'].mean())
    


working_data.head()
```

```python
working_data['order'] = working_data['production']
```

```python
alt.Chart(working_data).mark_area().encode(
    x = alt.X('date:T'),
    y = alt.Y('production:Q', aggregate = 'sum'),
    color = alt.Color('source:N'),
    order = alt.Order('order:N', aggregate='sum', sort = 'descending')
).properties(width = 1920, height = 300, background = '#F9F9F9', padding = 25)
```

```python
import altair as alt
from vega_datasets import data

source = data.barley()

alt.Chart(source).mark_bar().transform_calculate(
    key="datum.site == 'Crookston'"
).transform_joinaggregate(
    sort_key="argmax(key)", groupby=['variety']
).transform_calculate(
    sort_val='datum.sort_key.value'  
).encode(
    x=alt.X('sum(yield)', stack='normalize'),
    y=alt.Y('variety', sort=alt.SortField('sort_val', order='ascending')),
    color='site',
    order=alt.Order(
      # Sort the segments of the bars by this field
      'site',
      sort='ascending'
    )
)
```
